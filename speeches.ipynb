```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

nltk.download('all-nltk')   #popular, all-nltk

```

```python
text = open('speeches/2017-Trump.txt', 'r', encoding = 'latin-1').read()
text
```

```python
import re
text = re.sub(r"[^\w\s]+|\s+", " ", text).strip()
print(text)
```

```python
tokens = text.split()
tokens = [token.lower() for token in tokens if token.isalpha()]    #alphabet
print(tokens)
```

```python
freq = nltk.FreqDist(tokens)
freq
```

```python
stop_words = stopwords.words('english') 
print(stop_words)
```

```python
ctokens = tokens.copy()
for token in tokens:
    if token in stop_words:
        ctokens.remove(token)

print(ctokens)
```

```python
freq2 = nltk.FreqDist(ctokens)
freq2
```
